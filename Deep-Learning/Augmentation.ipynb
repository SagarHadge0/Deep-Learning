```python
image_string1 =tf.io.read_file(image_path)   #convert image to tensor
```

```python
image_string1
```

```python
image = tf.image.decode_jpeg(image_string1,channels=3)
```

```python
plt.imshow(image)
```

```python
def visualize(orinal,augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(orinal)
    plt.xticks([])
    plt.yticks([])
    plt.subplot(1,2,2)
    plt.title('Augmented')
    plt.imshow(augmented)
    plt.xticks([])
    plt.yticks([])
```

```python
flipped = tf.image.flip_left_right(image)
```

```python
plt.imshow(flipped)
```

```python
visualize(image,flipped)
```

```python
Gryscaled = tf.image.rgb_to_grayscale(image)
```

```python
visualize(image,tf.squeeze(Gryscaled))
plt.colorbar()
```

```python
plt.imshow(tf.squeeze(Gryscaled),cmap='gray_r')
```

```python
saturated = tf.image.adjust_saturation(image,5)
visualize(image,saturated)
```

```python
bright = tf.image.adjust_brightness(image,0.5)
visualize(image,bright)
```

```python
rotate = tf.image.rot90(image)
visualize(image,rotate)
```

```python
cencroped = tf.image.central_crop(image,central_fraction=0.5)
visualize(image,cencroped)
```

```python
contras = tf.image.adjust_contrast(image,0.9)
visualize(image,contras)
```
